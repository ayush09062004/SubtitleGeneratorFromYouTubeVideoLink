```python
# Step 1: Install required libraries
!pip install youtube-transcript-api

# Step 2: Import necessary libraries
from youtube_transcript_api import YouTubeTranscriptApi
import re

# Step 3: Function to extract video ID from the YouTube link
def get_video_id(youtube_url):
    # Regular expression to extract the video ID from the URL
    regex = r"(?:v=|\/)([0-9A-Za-z_-]{11}).*"
    match = re.search(regex, youtube_url)
    if match:
        return match.group(1)
    else:
        return None

# Step 4: Function to fetch subtitles from the YouTube video
def fetch_subtitles(youtube_url, language='en'):
    video_id = get_video_id(youtube_url)
    if video_id:
        try:
            # Fetch subtitles using YouTubeTranscriptApi
            transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[language])
            subtitles = ""
            for item in transcript:
                subtitles += f"{item['text']}\n"
            return subtitles
        except Exception as e:
            return f"Error fetching subtitles: {str(e)}"
    else:
        return "Invalid YouTube URL"

# Step 5: Provide the YouTube video URL
youtube_url = 'https://youtu.be/sZuy356qkPM?si=DWUWhvRSIXOJ_Za4'

# Fetch and display the subtitles
subtitles = fetch_subtitles(youtube_url)
print(subtitles)
```
